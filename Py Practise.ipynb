# Python game

The game we are going to monetize is going to considers a portion of a card deck including all the number cards from 1 to 10 and all the suits of clubs, diamonds, hearts and spades. For maintaing a consistency and improving readability we would be considering abbreviation of all the suits
- C for Clubs
- D for Diamonds
- H for Hearts
- S for Spades

We would be utilising statistics and random libraries for deriving statistical data of all the simulated results and to shuffle the cards respectively.

Statistics and Random libraries are imported. The deck of cards has been defined using abbreviation of all the suits. We are generating a list with all the required parameters. 

```python
# Importing required libraries
import statistics
import random
# Defining the structure with all the values
cards= [[2,'c'],[3,'c'],[4,'c'],[5,'c'],[6,'c'],[7,'c'],[8,'c'],[9,'c'],[10,'c'],
        [2,'d'],[3,'d'],[4,'d'],[5,'d'],[6,'d'],[7,'d'],[8,'d'],[9,'d'],[10,'d'],
        [2,'h'],[3,'h'],[4,'h'],[5,'h'],[6,'h'],[7,'h'],[8,'h'],[9,'h'],[10,'h'],
        [2,'s'],[3,'s'],[4,'s'],[5,'s'],[6,'s'],[7,'s'],[8,'s'],[9,'s'],[10,'s']]
print(cards)
```

Here we are calculating the deck total by doing addition of all the values of the suits.

```python
# Calculating the total value of all cards
i = 0
total = 0
while i <= 35:
    total = total + cards[i][0]
    i += 1
print('Total value of all cards: ',total)
```

The sum_of_cards function is used to calculate the total points of each player playing the game. The shuffle function is used to randomly shuffle the cards and assign it to the players in the game.

```python
# Function to calculate the sum of cards in a deck
def sum_of_cards(deck):
    i = 0
    total = 0
    while i <= 17:
        total = total + deck[i][0]
        i += 1
    return total

# Function to shuffle the cards
def shuffle(cards):
    player_one = list(random.sample(cards, k = 18))
    player_two = list(i for i in cards if i not in player_one)
    return player_one, player_two

# Calling the shuffle card function and assigning to variables
shuffled_set = shuffle(cards)
player_one_deck = shuffled_set[0]
player_two_deck = shuffled_set[1]
# Printing all the required values
print('Cards dealt to player one\n', player_one_deck)
print('\nCards dealt to player two\n', player_two_deck)
print('\nNo of cards dealt to player one is', len(shuffled_set[0]))
print('\nNo of cards dealt to player two is', len(shuffled_set[1]))
print('\nSum of all cards dealt to player one is', sum_of_cards(player_one_deck))
print('\nSum of all cards dealt to player two is', sum_of_cards(player_two_deck))
```

The game function is the heart of the entire operation. It would take into consideration the first two cards in the deck of each player and compare the sum of their values. The player with the lower sum of values would be the winner of the round and we would proceed to the next rounds. Whenever a player loses the round, the cards in that round would be added to the losing players deck. Finally when a player runs out of a card he would be declared the winner or if 50 round concludes and the player with the lowest sum of scores of all cards would be declared the winner.

```python
# Function to simulate the game
def game(player_one_deck, player_two_deck):
    player_one_total = sum_of_cards(player_one_deck)
    player_two_total = sum_of_cards(player_two_deck)
    i = 0
    # Loop to run the game till player run out of cards or 50 round concludes
    while i <= 50 and player_one_deck and player_two_deck:
        player1_game_deck = player_one_deck[:2]
        print('\nPlayer one cards for this round \n', player1_game_deck)
        player_one_deck = player_one_deck[2:]
        print('\nPlayer one rest of cards \n', player_one_deck)
        
        player2_game_deck = player_two_deck[:2]
        print('\nPlayer two cards for this round \n', player2_game_deck)
        player_two_deck = player_two_deck[2:]
        print('\nPlayer two rest of cards \n', player_two_deck)
        
        while True:
            # Calculating total for each round and comparing it for both players and extending the deck of
            # player who loses the round
            player1_total = sum(d[0] for d in player1_game_deck)
            player2_total = sum(d[0] for d in player2_game_deck)
            if player1_total < player2_total:
                print('Player 1 wins the round')
                player_two_deck.extend(player1_game_deck)
                player_two_deck.extend(player2_game_deck)
                break
            elif player2_total < player1_total:
                print('Player 2 wins the round')
                player_one_deck.extend(player2_game_deck)
                player_one_deck.extend(player1_game_deck)
                break
            else:
                print('The round was a draw')
                if not player_one_deck or not player_two_deck:
                    break  
                player1_game_deck.extend(player_one_deck[:2])
                player2_game_deck.extend(player_two_deck[:2])
                player_one_deck = player_one_deck[2:]
                player_two_deck = player_two_deck[2:]            
        i += 1
    # Conditional checks to find the winner of the game
    if not player_one_deck:
        print('\nPlayer one wins the game by getting rid of all the cards')
        return sum(d[0] for d in player_one_deck), sum(d[0] for d in player_two_deck), player_one_total
    elif not player_two_deck:
        print('\nPlayer two wins the game by getting rid of all the cards')
        return sum(d[0] for d in player_two_deck), sum(d[0] for d in player_one_deck), player_two_total
    else:
        if sum(d[0] for d in player_one_deck) > sum(d[0] for d in player_two_deck):
            print('\nPlayer 2 wins the game after 50 rounds with lower sum of total cards')
            return sum(d[0] for d in player_two_deck), sum(d[0] for d in player_one_deck), player_two_total
        elif sum(d[0] for d in player_one_deck) < sum(d[0] for d in player_two_deck):
            print('\nPlayer 1 wins the game after 50 rounds with lower sum of total cards')
            return sum(d[0] for d in player_one_deck), sum(d[0] for d in player_two_deck), player_one_total
        else:
            print('\nThe scores are still equal after 50 rounds')
            return sum(d[0] for d in player_one_deck), sum(d[0] for d in player_two_deck), 0    
game(player_one_deck, player_two_deck)
```

Below we are making using of the statistics library to calculate the Maximum score with which a player won the game, Maximum score with which a player lost the game, Minimum score with which a player won the game, Minimum score with which a player lost the game, Mean score by which players won the game, Mean score by which players lost the game, The maximum start total of deck of a winner. We have also created a dummy function of the main game function without all the print statements.

```python
# Dummy function to simulate the game without printing anything
def game_without_print(player_one_deck, player_two_deck):
    player_one_total = sum_of_cards(player_one_deck)
    player_two_total = sum_of_cards(player_two_deck)
    i = 0
    while i <= 50 and player_one_deck and player_two_deck:
        player1_game_deck = player_one_deck[:2]
        player_one_deck = player_one_deck[2:]
        
        player2_game_deck = player_two_deck[:2]
        player_two_deck = player_two_deck[2:]
        
        while True:
            player1_total = sum(d[0] for d in player1_game_deck)
            player2_total = sum(d[0] for d in player2_game_deck)
            if player1_total < player2_total:
                player_two_deck.extend(player1_game_deck)
                player_two_deck.extend(player2_game_deck)
                break
            elif player2_total < player1_total:
                player_one_deck.extend(player2_game_deck)
                player_one_deck.extend(player1_game_deck)
                break
            else:
                if not player_one_deck or not player_two_deck:
                    break  
                player1_game_deck.extend(player_one_deck[:2])
                player2_game_deck.extend(player_two_deck[:2])
                player_one_deck = player_one_deck[2:]
                player_two_deck = player_two_deck[2:]            
        i += 1
    if not player_one_deck:
        return sum(d[0] for d in player_one_deck), sum(d[0] for d in player_two_deck), player_one_total
    elif not player_two_deck:
        return sum(d[0] for d in player_two_deck), sum(d[0] for d in player_one_deck), player_two_total
    else:
        if sum(d[0] for d in player_one_deck) > sum(d[0] for d in player_two_deck):
            return sum(d[0] for d in player_two_deck), sum(d[0] for d in player_one_deck), player_two_total
        elif sum(d[0] for d in player_one_deck) < sum(d[0] for d in player_two_deck):
            return sum(d[0] for d in player_one_deck), sum(d[0] for d in player_two_deck), player_one_total
        else:
            return sum(d[0] for d in player_one_deck), sum(d[0] for d in player_two_deck), 0

# Function to perform statistical analysis
def game_stats(runs):
    winning_scores = []
    losing_scores = []
    max_starting_score = []
    for i in range(runs):
        shuffled_set = shuffle(cards)
        player_one_deck = shuffled_set[0]
        player_two_deck = shuffled_set[1]
        winner_score, loser_score, max_score_start = game_without_print(player_one_deck, player_two_deck)
        winning_scores.append(winner_score)
        losing_scores.append(loser_score)
        max_starting_score.append(max_score_start)
    return winning_scores, losing_scores, max_starting_score

no_of_runs = 1000
winning_scores, losing_scores, max_start = game_stats(no_of_runs)
# Calculating all the statistical values based on the list of scores
max_score_win = max(winning_scores)
max_score_lose = max(losing_scores)
min_score_win = min(winning_scores)
min_score_lose = min(losing_scores)
mean_score_win = statistics.mean(winning_scores)
mean_score_lose = statistics.mean(losing_scores)
maximum_start_score = max(max_start)
# Printing all the values based on the statistics
print('Maximum score with which a player won the game', max_score_win)
print('Maximum score with which a player lost the game', max_score_lose)
print('Minimum score with which a player won the game', min_score_win)
print('Minimum score with which a player lost the game', min_score_lose)
print('Mean score by which players won the game', mean_score_win)
print('Mean score by which players lost the game', mean_score_lose)
print('The maximum start total of deck of a winner is', maximum_start_score)

win_more_points = len([i for i in max_start if i > 108])
print(f'Percentage of games won by a player with higher total points compared to the opponent {(win_more_points/ no_of_runs) * 100:.2f}%')




        
    
    
```
