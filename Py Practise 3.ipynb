## Data Cleaning of INC5000 data
We need to perform cleaning of data in the dataset to iron out any inconsistency or redundant data.

We could primarily see 5 key issues in the data which needs to be fixed. They are as follows :-

- The REVENUE column is misspelled
- The STATE CODE column is incorrectly name STATE
- The STATE column is incorrectly named CITY
- The dataset has two redundant columns at the end with information not required for our analysis
- We already have the column for both CITY and STATE, so an extra column with both together is redundant

To proceed with further analysis, we proceed to implement necessary data manipulation to fix these issues.

```python
# Implementing necessary data cleaning on the data set
# Fixing the column named Revenue
inc_cleaned = inc.rename(columns={'REVENUW': 'REVENUE','STATE':'STATE CODE','CITY.1':'STATE'})
```

```python
# Dropping columns that are redundant for the overall data set
inc_cleaned.drop(['Unnamed: 14', 'Unnamed: 15','CITY, STATE'], axis=1, inplace=True)
```

```python
# Displaying the dataframe after implementing cleaning
inc_cleaned
```

## Loading required region data for further analysis
We don't have the region data to further proceed with our analysis, but we have another dataset which has the necessary information that is needed. It has all the states by their region which would be utilized for upcoming analysis.

The file we would be using is in CSV format, so we would be using a function from pandas library to load it into a data frame.

```python
#Loading new data set which contains region data for analysis
states_region = pd.read_csv('HW4_states_by_region.csv')
```

## Merging and further data cleaning for region mapping
We need to merge the region column onto our dataframe containing all the information of the companies, so we could see which state belongs to which region. Pandas provide multiple ways of merging two dataframes. They include

- left: use only keys from left frame, similar to a SQL left outer join; preserve key order.

- right: use only keys from right frame, similar to a SQL right outer join; preserve key order.

- outer: use union of keys from both frames, similar to a SQL full outer join; sort keys lexicographically.

- inner: use intersection of keys from both frames, similar to a SQL inner join; preserve the order of the left keys.

- cross: creates the cartesian product from both frames, preserves the order of the left keys.

We would be using left merge and mapping the Key 'STATE' of inc dataset with key 'state' of the state region dataset.

We also need to do some data cleaning to fix some inconsistencies and remove unwanted columns. The changes we would be implementing include

- Capitalizing the name of the column 'region' for consistency across the column names.

- Dropping the redundant columns of State, State Code and Division that were formed during the merge.

We can finally get a consistend dataframe on which merging has been performed and data has been cleaned.

```python
# Performing the merge using left merge
inc_with_region = inc_cleaned.merge(states_region, left_on='STATE', right_on='State', how='left')

# Capitalizing the column region for consistency in the inc dataframe
inc_with_region.rename(columns={'Region': 'REGION'}, inplace=True)

# Dropping redundant columns not required for our analysis
inc_with_region.drop(['State', 'State Code','Division'], axis=1, inplace=True)
```

```python
# Displaying the dataframe after all cleaning and merging, with the REGION column
inc_with_region
```

## Function to filter based on categories and sort based on revenue
We need to create a function which takes the following parameters as input

- Dataframe of INC5000 data.

- Category to which the company belongs to.

- The value of the Category field needed to be matched.

- A constant that defines the number of rows to be displayed.

The function returns the dataframe after applying necessary filters and sorts the dataframe in descending order of the revenues from highest to lowest.

This function could help us see which are the top performers across various parameters such as city, state, industry. 

```python
# Function to take parameters of data frame, category, value, and a number that specifies the number of rows 
# to display in the result. The function should return a data frame containing the top n
# rows of companies, based on revenues for a specific value in the category, only if both category and
# value are provided
def company_sorter(dataframe, category= 'None', value= 'None', rows = 10):
    if category != 'None' and value != 'None':
        category_by_revenue = dataframe[dataframe[category] == value]
        sorted_category_by_revenue = category_by_revenue.nlargest(rows, 'REVENUE')
        return sorted_category_by_revenue
    
    # If both category and value are not provided, the top n companies overall should be
    # displayed 
    else:
        sorted_category_by_revenue = dataframe.nlargest(rows, 'REVENUE')
        return sorted_category_by_revenue
```

### We are displaying the top 15 companies irrespective of the industry the belong to or the location where they are based. Basically we are not specifying any category and getting a list of companies in the descending order of their revenues.

```python
# Displaying the top fifteen overall companies
top_fifteen = company_sorter(inc_with_region, rows = 15)
top_fifteen
```

### We are displaying the top performing companies by revenue in the Food and Beverage industry and displaying 10 rows of data that is companies that belong in top 10 in that particular sector.

```python
# Displaying the top ten companies in the food and beverage industry
top_food_beverage = company_sorter(inc_with_region, 'INDUSTRY', 'Food & Beverage', 10)
top_food_beverage
```

### We are displaying the top 3 companies which are located in the city of New york.

```python
# Displaying the top three companies located in New York City
top_nyc = company_sorter(inc_with_region, 'CITY', 'New York City', 3)
top_nyc
```

### We are displaying top five companies that are situated in the State of Maryland

```python
# Displaying the top five companies in Maryland
top_maryland = company_sorter(inc_with_region, 'STATE', 'Maryland', 5)
top_maryland
```

## Assigning awards to top performing companies with word 'Services' in company name
We are required to assign awards to companies that have the word Services in their industry column and are top performers in the terms of revenue. The top quarter of the companies would receive the Trailblazer and the remaining companies in the top half on basis of revenue would receive the Pioneer award.

We would need to first filter out and create a dataframe with companies which have the word 'Services' in their Industry column.

We need to sort all the data

Then we would calculate the number of companies in the top quarter and top half respectively.

We would then create a new column to assign the awards to the respective companies on basis of where they lie in terms of their revenue.

We would assign the value Trailblazer to all companies in the top quarter and assign the value Pioneer to rest of the companies in the top half.

This column would help us comprhend where a particular company stands and in which quartile does the company belong to in terms of revenue.

```python
# Making a new dataframe with all industries that contains services in their name
service_industries = inc_with_region[inc_with_region['INDUSTRY'].str.contains('Services')]
```

```python
# Finding the value of top quarter of all service based industries
top_quarter = round((len(service_industries)/100) * 25)

# Finding the value of top half of all service based industries
top_half = round((len(service_industries)/100) * 50)
```

```python
# Sorting the dataframe by descending order of their revenues
service_sorted_by_revenue = service_industries.sort_values('REVENUE', ascending=False)
```

```python
# Displaying the new dataframe of all service based industries in descending order of their revenues
service_sorted_by_revenue
```

```python
# Creating a dataframe of companies in the top quarter
Trailblazer = service_sorted_by_revenue[0:top_quarter]
Trailblazer = Trailblazer.copy()

# Creating a new column award in the Trailblazer dataframe and populating it with name of the award
Trailblazer.loc[:,'AWARD'] = 'Trailblazer'
Trailblazer
```

```python
# Creating a dataframe of companies in the top half excluding all in the top quarter
Pioneer = service_sorted_by_revenue[top_quarter:top_half]
Pioneer = Pioneer.copy()

# Creating a new column award in the Pioneer dataframe and populating it with name of the award
Pioneer.loc[:,'AWARD'] = 'Pioneer'
Pioneer
```

We are also performing necessary cleaning on the dataframe to remove the columns that we don't require. Essentially, excluding Rank, Company Name, Revenue and Award all other columns are not required. 

```python
# Concating the result of both type of companies with awards and creating a new dataframe
awarded_companies = pd.concat([Trailblazer,Pioneer])

# Performing data cleaning on the dataframe and removing all unwanted columns
awarded_companies.drop(['URL', 'CITY','GROWTH', 'EMPLOYEES', 'WEBSITE', 'STATE', 'STATE CODE',
                        'ZIP CODE', 'FOUNDED', 'INDUSTRY', 'REGION'], axis=1, inplace=True)

awarded_companies
```

## Additional Analysis
We need to look through the various states across US which have companies with highest revenue and also calculate the number of companies within each state. We would like to visualize the data and make inferences on basis of the available data.

```python
# New data frame with State and sum of revenues of all companies in that state and the count of companies
all_states = inc_with_region.groupby('STATE').agg({'REVENUE': 'sum', 'COMPANY NAME': 'count'})
```

```python
all_states
```

```python
# Sorting the data with descending order of revenue
all_states = all_states.sort_values(by='REVENUE', ascending=False)
```

```python
# Printing the top 5 states with highest total revenue generated by companies
print(all_states.head())
```

```python
# Isolating the state with the highest total revenue and displaying the total revenue and number of companies
state_highest_revenue = all_states.iloc[0]
print('State with the Highest Revenue:', state_highest_revenue.name)
print('Total Revenue in the Highest Revenue State:', state_highest_revenue['REVENUE'])
print('Number of Companies in the Highest Revenue State:', state_highest_revenue['COMPANY NAME'])
```

```python
# Importing libraries for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Plotting a bar graph with necessary parameters using the imported libraries
# Providing appropriate title and labels.
sns.set(style='whitegrid')
fig, ax = plt.subplots(figsize=(14, 8))
sns.barplot(x=all_states.head(10).index, y=all_states.head(10)['REVENUE'], ax=ax, palette='viridis')
ax.set_title('Top 10 States with Highest Revenue')
ax.set_xlabel('State')
ax.set_ylabel('Total Revenue in billions')
plt.tight_layout()
plt.show()
```
